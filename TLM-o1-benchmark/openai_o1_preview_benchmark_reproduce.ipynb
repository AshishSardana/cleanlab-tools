```python
import pandas as pd

# Here we use the PII-98 dataset as an example. But you can use any other dataset such as SVAMP-49 or TriviaQA-114.
pii_98_dataset = pd.read_csv("PII-98-benchmark/pii_98_dataset.csv")
pii_98_dataset.head()
```

```python
from openai import OpenAI

client = OpenAI(api_key=open_ai_key)

responses = []
# Only use the first 5 rows for illustration. Feel free to use the entire dataset.
for _, row in pii_98_dataset.head().iterrows():
    prompt = row['prompt']
    response = client.chat.completions.create(
        messages=[{"role": "user", "content": prompt}],
        model="o1-preview",
    )
    response_text = response.choices[0].message.content
    responses.append(response_text)
    print(f"Prompt: {prompt}")
    print(f"Response: {response_text}")
    print(f"Ground Truth: {row['ground_truth']}")
    print("---")

# Create a new DataFrame with the responses
results_df = pd.DataFrame({
    'prompt': pii_98_dataset.head()['prompt'],
    'ground_truth': pii_98_dataset.head()['ground_truth'],
    'response_openai_o1_preview': responses
})

results_df.to_csv("PII-98-benchmark/pii_98_openai_o1_preview_responses_new.csv", index=False)

# Display the results
results_df
```
