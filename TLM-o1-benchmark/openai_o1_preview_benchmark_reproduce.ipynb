```python
import pandas as pd

# Here we use the SVAMP-49 dataset as an example. But you can use any other dataset such as PII-98 or TriviaQA-114.
svamp_49_dataset = pd.read_csv("SVAMP-49-benchmark/svamp_49_dataset.csv")
svamp_49_dataset.head()
```

```python
from openai import OpenAI

open_ai_key = "<YOUR OPEN AI KEY HERE>"
client = OpenAI(api_key=open_ai_key)

responses = []
# Only use the first 5 rows for illustration. Feel free to use the entire dataset.
for _, row in svamp_49_dataset.head().iterrows():
    prompt = row['prompt']
    response = client.chat.completions.create(
        messages=[{"role": "user", "content": prompt}],
        model="o1-preview",
    )
    response_text = response.choices[0].message.content
    responses.append(response_text)
    print(f"Prompt: {prompt}")
    print(f"Response: {response_text}")
    print(f"Ground Truth: {row['ground_truth']}")
    print("---")

# Create a new DataFrame with the responses
results_df = pd.DataFrame({
    'prompt': svamp_49_dataset.head()['prompt'],
    'ground_truth': svamp_49_dataset.head()['ground_truth'],
    'response_openai_o1_preview': responses
})

results_df.to_csv("SVAMP-49-benchmark/svamp_49_openai_o1_preview_responses_new.csv", index=False)

# Display the results
results_df
```
