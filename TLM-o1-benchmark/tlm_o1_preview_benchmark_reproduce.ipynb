```python
from cleanlab_studio import Studio

studio = Studio("<your-api-key>")
```

```python
import pandas as pd

# Here we use the PII-98 dataset as an example. But you can use any other dataset such as SVAMP-49 or TriviaQA-114.
pii_98_dataset = pd.read_csv("PII-98-benchmark/pii_98_dataset.csv")
pii_98_dataset.head()
```

```python
tlm = studio.TLM(options={"model": "o1-preview"}, quality_preset="best")

responses = []
trustworthiness_scores = []
# Only use the first 5 rows for illustration. Feel free to use the entire dataset.
for _, row in pii_98_dataset.head().iterrows():
    prompt = row['prompt']
    response_dict = tlm.prompt(prompt)
    responses.append(response_dict['response'])
    trustworthiness_scores.append(response_dict['trustworthiness_score'])
    print(f"Prompt: {prompt}")
    print(f"Response: {response_dict['response']}")
    print(f"Trustworthiness Score: {response_dict['trustworthiness_score']}")
    print(f"Ground Truth: {row['ground_truth']}")
    print("---")

# Create a new DataFrame with the responses and trustworthiness scores
results_df = pd.DataFrame({
    'prompt': pii_98_dataset.head()['prompt'],
    'ground_truth': pii_98_dataset.head()['ground_truth'],
    'response_tlm_o1_preview': responses,
    'trustworthiness_score': trustworthiness_scores
})

results_df.to_csv("PII-98-benchmark/pii_98_tlm_o1_preview_responses_new.csv", index=False)

# Display the results
results_df
```
